```python
import numpy
import matplotlib.pyplot as plt
```

```python
from numpy import genfromtxt

dev_data = genfromtxt('run_summary_RB12_LR0.0001_ENS0.2_CON0.5_PHY0.125_dev-tag-Generator_loss.csv', delimiter=',')
dev_step = dev_data[1:,1]
dev_loss = dev_data[1:,2]

train_data = genfromtxt('run_summary_RB12_LR0.0001_ENS0.2_CON0.5_PHY0.125_train-tag-Generator_loss.csv', delimiter=',')
train_step = train_data[1:,1]
train_loss = train_data[1:,2]
```

```python
plt.figure(figsize=(10,6))
plt.plot(dev_step, dev_loss, label='dev')
plt.plot(train_step, train_loss, label='train')
plt.xlabel('step', fontsize=16)
plt.ylabel('generator loss', fontsize=16)
plt.legend(fontsize=14)
plt.title('TEResNet',fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.semilogy(dev_step, dev_loss, label='dev')
plt.semilogy(train_step, train_loss, label='train')
plt.xlabel('step', fontsize=16)
plt.ylabel('generator loss', fontsize=16)
plt.legend(fontsize=14)
plt.title('TEResNet',fontsize=16)
plt.show()
```
